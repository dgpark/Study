### API 사용하여 데이터 저장

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

customerData = {}

agePrdL = [] #나이대
crdCoBzkNmL = [] # 업종명칭
crdUseAmtL = [] # 사용금액
crdUseCntL = [] # 사용건수
fstCtgyL = [] #1차분류 코드
fstCtgyNmL = []#1차분류 코드명
noL = [] # 일련번호
sndCtgyL = [] #2차분류 코드
sndCtgyNmL = [] #2차분류 코드명
sxClNmL = [] #성별
useYymmL = [] #사용연월

for page in range(1,200, 10):
    url = 'http://apis.data.go.kr/B552640/GwpCardUseDtlInfoService/selectListUseDtl?serviceKey=Xai3KLBM%2BwQZuaih0gOwxP2tblAZoVHdUTID07%2BUGOEGv7oU%2BTKKmd5naEPitH51Hqj1Zd%2FIyQ7mYv8UmcmXvw%3D%3D&numOfRows=10&pageNo={}&useYymm=201801&sxClNm=%EB%82%A8'.format(page)
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')

    agePrd = soup.find_all('ageprd')
    crdCoBzkNm = soup.find_all('crdcobzknm')
    crdUseAmt = soup.find_all('crduseamt')
    crdUseCnt = soup.find_all('crdusecnt')
    fstCtgy = soup.find_all('fstctgy')
    fstCtgyNm = soup.find_all('fstctgynm')

    no = soup.find_all('no')
    sndCtgy = soup.find_all('sndctgy')
    sndCtgyNm = soup.find_all('sndctgynm')
    sxClNm = soup.find_all('sxclnm')
    useYymm = soup.find_all('useyymm')

    for a in agePrd :
        agePrdL.append(a.text)
    for ccn in crdCoBzkNm:
        crdCoBzkNmL.append(ccn.text)
    for cua in crdUseAmt:
        crdUseAmtL.append(cua.text)
    for cuc in crdUseCnt:
        crdUseCntL.append(cuc.text)
    for fc in fstCtgy:
        fstCtgyL.append(fc.text)
    for fcn in fstCtgyNm:
        fstCtgyNmL.append(fcn.text)

    for n in no:
        noL.append(n.text)
    for sc in sndCtgy:
        sndCtgyL.append(sc.text)
    for scn in sndCtgyNm:
        sndCtgyNmL.append(scn.text)
    for scn in sxClNm:
        sxClNmL.append(scn.text)
    for uY in useYymm:
        useYymmL.append(uY.text)    

print(agePrdL[0]) #나이대
print(crdCoBzkNmL[0]) # 업종명칭
print(crdUseAmtL[0]) # 사용금액
print(crdUseCntL[0]) # 사용건수
print(fstCtgyL[0]) #1차분류 코드
print(fstCtgyNmL[0])#1차분류 코드명
print(noL[0]) # 일련번호
print(sndCtgyL[0]) #2차분류 코드
print(sndCtgyNmL[0]) #2차분류 코드명
print(sxClNmL[0]) #성별
print(useYymmL[0]) #사용연월
```
